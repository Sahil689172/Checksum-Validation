```python
from scapy.all import *

# Load your capture file
packets = rdpcap("network_capture.pcap")

print("Total packets:", len(packets))
print("-" * 60)

ip_valid = 0
tcp_valid = 0

for i, pkt in enumerate(packets):

    # Work only with IP packets
    if IP in pkt:

        print(f"\nPacket No: {i+1}")

        # Store original checksums
        original_ip = pkt[IP].chksum

        # Delete checksum so Scapy recalculates
        del pkt[IP].chksum

        # Rebuild packet
        rebuilt_pkt = IP(bytes(pkt[IP]))
        recalculated_ip = rebuilt_pkt.chksum

        print("Source IP      :", pkt[IP].src)
        print("Destination IP :", pkt[IP].dst)

        print("Original IP Checksum   :", hex(original_ip))
        print("Calculated IP Checksum :", hex(recalculated_ip))

        if original_ip == recalculated_ip:
            print("IP Checksum VERIFIED ✅")
            ip_valid += 1
        else:
            print("IP Checksum MISMATCH ❌")

        # ---------- TCP VALIDATION ----------
        if TCP in rebuilt_pkt:

            original_tcp = pkt[TCP].chksum

            # Delete TCP checksum
            del rebuilt_pkt[TCP].chksum

            # Rebuild again (important!)
            final_pkt = IP(bytes(rebuilt_pkt))
            recalculated_tcp = final_pkt[TCP].chksum

            print("\nOriginal TCP Checksum   :", hex(original_tcp))
            print("Calculated TCP Checksum :", hex(recalculated_tcp))

            if original_tcp == recalculated_tcp:
                print("TCP Checksum VERIFIED ✅")
                tcp_valid += 1
            else:
                print("TCP Checksum MISMATCH ❌")

        print("-" * 60)


print("\n========= FINAL REPORT =========")
print("Valid IP Checksums  :", ip_valid)
print("Valid TCP Checksums :", tcp_valid)
```
